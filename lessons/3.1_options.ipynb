## Part 1: Risk-Neutral Probabilities

- Let $S_0, S_1, \dots, S_n$ be a sequence of random variables denoting the evolution of a stock's price.

- Consider a general one-period binomial model. Then:

$$S_0 \cdot u = S_1(H) \quad \text{and} \quad S_0 \cdot d = S_1(T)$$

- We define $u$ as an up factor and $d$ as a down factor, such that:
$$u = \frac{S_1(H)}{S_0} \quad \text{and} \quad d = \frac{S_1(T)}{S_0}$$

- Also, let $r$ denote the rate of interest.

    - One dollar invested in the money market at time zero will yield $1+r$ dollars at time one.
    - One dollar borrowed from the money market will result in a debt of $1+r$ dollars at time one.

- **Definition: Arbitrage**

    - A trading strategy that begins with no money, has zero probability of losing money, and has a positive probability of making money.
    - **Notes**:
        - A mathematical model that allows for arbitrage cannot be used for pricing analysis.
        - Real markets typically exhibit arbitrage, but due to competition, it is short-lived.

- In the binomial asset-pricing model, to rule out arbitrage, we must assume: $$0 < d < 1 + r < u$$
    
    - $d > 0$ because stock prices must remain positive.
    - $d < 1 + r$: If this were not true, one could borrow the stock, pay $1 + r$ on the capital, and yield at least $d$, fitting the definition of arbitrage.
    - $u > 1 + r$: If this were not true, one could sell the stock short and invest the proceeds in the money market to achieve an arbitrage.

- **Risk-Neutral Measure**

    - We define the risk-neutral probabilities $\tilde{p}$ and $\tilde{q}$: $$\tilde{p} = \frac{(1+r) - d}{u - d}, \quad \tilde{q} = \frac{u - (1+r)}{u - d}$$

## Part 2: Conditional Expectation

- **Conditional Expectation**

    - Classical conditional expectation notation:

$$\mathbb{E}[X | Y = y] = \sum_{x}x \cdot \mathbb{P}[X = x | Y = y]$$

- **The Conditional Expectation of $S_{n+1}$**

    - The conditional expectation can be regarded as an estimate of the value of $S_{n+1}$ based on the knowledge of the first $n$ steps in the process.

    - Let $\tilde{p}$ denote the risk-neutral probability of an upward price movement, $\tilde{q}$ be the risk-neutral probability of a downward price movement, and let $\omega_1 \ldots \omega_n$ be fixed. Then, the conditional expectation of $S_{n+1}$ is:
    
    $$\tilde{\mathbb{E}}_n[S_{n+1}](\omega_1 \ldots \omega_n) = \frac{1}{1 + r}[\tilde{p}S_{n+1}(\omega_1 \ldots \omega_n H) + \tilde{q}S_{n+1}(\omega_1 \ldots \omega_n T)]$$

- **Definition: Conditional Expectation**

    - Let $n$ satisfy $1 \leq n \leq N$ and let $\omega_1 \ldots \omega_n$ be given and, for the moment, fixed. There are $2^{N-n}$ possible continuations $\omega_{n+1} \ldots \omega_N$ of the fixed sequence $\omega_1 \ldots \omega_n$.

    - Let $h$ denote the number of heads in the continuation $\omega_{n+1} \ldots \omega_N$, and let $t$ denote the number of tails in the continuation $\omega_{n+1} \ldots \omega_N$.

    - We define: 
    
    $$\tilde{\mathbb{E}}_n[X](\omega_1 \ldots \omega_n) = \sum_{\omega_{n+1} \ldots \omega_N} (\tilde{p})^h (\tilde{q})^t \cdot X(\omega_1 \ldots \omega_n \omega_{n+1} \ldots \omega_N)$$

    - In a sense, the sequence $\omega_1 \ldots \omega_n$ is our known price path (past stock price evolution), and the sequence $\omega_{n+1} \ldots \omega_N$ is our unknown price path (future stock price evolution).

    - It is considered a conditional expectation since we are _conditioning_ the expectation on the given information from the fixed sequence we have observed as $\omega_1 \ldots \omega_n$.

- **Theorem: Fundamental Properties of Conditional Expectations**

    - Let $N$ be a positive integer, and let $X$ and $Y$ be random variables depending on the first $N$ coin tosses. Let $1 \leq n \leq N$ be given. The following properties hold:

    1. **Linearity**: For all constants $c_1$ and $c_2$, we have: 
    $$\tilde{\mathbb{E}}_n[c_1X+c_2Y] = c_1\tilde{\mathbb{E}}_n[X] + c_2\tilde{\mathbb{E}}_n[Y]$$

    2. **What is Known**: If $X$ depends only on the first $n$ steps, then: 
    $$\tilde{\mathbb{E}}_n[XY] = X \tilde{\mathbb{E}}_n[Y]$$

    3. **Iterated Conditioning**: If $1 \leq n \leq m \leq N$, then: 
    $$\tilde{\mathbb{E}}_n[\tilde{\mathbb{E}}_m[X]] = \tilde{\mathbb{E}}_n[X]$$

    4. **Independence**: If $X$ depends only on tosses $n+1$ through $N$, then: 
    $$\tilde{\mathbb{E}}_n[X] = \tilde{\mathbb{E}}[X]$$

    - For example, consider $X(\omega_1 \ldots \omega_n, \omega_{n+1} \ldots \omega_N)$. $X$ is independent if $X$ does not depend on the first part of the sequence $\omega_1 \ldots \omega_n$.

## Part 3: Martingales

- Under the risk-neutral measure, for a stock that pays no dividend, the best estimate based on the information at time $n$ of the discounted value of the stock price at time $n + 1$ is the stock price at time $n$. Risk-neutral probabilities are selected to enforce this fact. Processes that satisfy this condition are called martingales.

- **Definition: Martingale**

    - Consider the binomial asset pricing model. Let $M_0, M_1, \dots, M_N$ be a sequence of random variables, with each $M_n$ depending on the first $n$ coin tosses (and $M_0$ constant). Such a sequence of random variables is called an adapted stochastic process.

        1. **Martingale** $$\tilde{\mathbb{E}}_n[M_{n+1}] = M_n, \quad n = 0, 1, \dots, N - 1$$

        2. **Submartingale** $$\tilde{\mathbb{E}}_n[M_{n+1}] \geq M_n, \quad n = 0, 1, \dots, N - 1$$

        3. **Supermartingale** $$\tilde{\mathbb{E}}_n[M_{n+1}] \leq M_n, \quad n = 0, 1, \dots, N - 1$$

- **Properties of Martingales**

    - **Multi-Step Property**

        - Whenever $0 \leq n \leq m \leq N$, $$\tilde{\mathbb{E}}_n[M_m] = M_n$$

    - **Constant Expectation Property**

        - The expectation of a martingale is constant over time, or, $$\tilde{\mathbb{E}}[M_n] = M_0, \quad n = 0, 1, 2, \dots, N$$

- Conceptually, a martingale represents a _fair game_, where your wealth may flucuate at each step, but overall it will average out to what you started with

## Part 3: Martingales (cont'd)

- **Theorem: Risk-Neutral Pricing Formula**

    - Consider an $N$-period binomial asset-pricing model with $0 < d < 1 + r < u$ and with risk-neutral probability measure $\tilde{\mathbb{P}}$. Let $V_N$ be a random variable (a derivative security paying off at time $N$) depending on the coin tosses. Then, for $n$ between $0$ and $N$, the price of the derivative security at time $n$ is given by the risk-neutral pricing formula:
    
    $$V_n = \tilde{\mathbb{E}}_n\left[\frac{V_N}{(1+r)^{N-n}}\right]$$

    - Furthermore, the discounted price of the derivative security is a martingale under $\tilde{\mathbb{P}}$:
    
    $$\frac{V_n}{(1+r)^n} = \tilde{\mathbb{E}}_n\left[\frac{V_{n+1}}{(1+r)^{n+1}}\right], \quad n = 0, 1, \dots, N - 1$$

    - This model can be used to represent derivative instruments such as European call or put options

- **Theorem: Cash Flow Valuation**

    - Consider an $N$-period binomial asset-pricing model with $0 < d < 1 + r < u$, and with risk-neutral probability measure $\tilde{\mathbb{P}}$. Let $C_0, C_1, \dots, C_N$ be a sequence of random variables such that each $C_n$ depends only on $w_1, \dots, w_n$. The price at time $n$ of the derivative security that makes payments $C_n, \dots, C_N$ at times $n, \dots, N$ respectively, is:
    
    $$V_n = \tilde{\mathbb{E}}_n\left[\sum_{k = n}^{N} \frac{C_k}{(1+r)^{k-n}}\right], \quad n = 0, 1, \dots, N$$

    - This model can be used to represent products with multiple cash flows, such as swaps, or bonds that carry embedded options

## Part 4: Markov Chains

- A Markov Chain is a sequence of random variables in discrete time, typically denoted by $X_n$, where $n$ is the index of the random variable at the $n$-th step.

- A Markov process (which is a type of stochastic process) is a sequence of random variables in continuous time, typically denoted by $X(t)$, where $t$ is a parameter representing time.

- **The Markov Property**

    - The Markov property states that the future state of a process depends only on the current state, and not on the sequence of events that preceded it. Formally, for any $n$,
    
    $$\mathbb{P}(X_{n+1} = x_{n+1} | X_0 = x_0, X_1 = x_1, \ldots, X_n = x_n) = \mathbb{P}(X_{n+1} = x_{n+1} | X_n = x_n)$$

    - In the context of conditional expectation, if $X$ is a Markov process, then:
    
    $$\mathbb{E}_n[f(X_{n+1})] = \mathbb{E}[f(X_{n+1}) | X_n]$$

- **Definition: Markov Process (Binomial Asset-Pricing Theory)**

    - Consider the binomial asset-pricing model. Let $X_0, X_1, \ldots, X_N$ be an adapted process. If, for every $n$ between $0$ and $N-1$ and for every function $f(x)$, there is another function $g(x)$ (depending on $n$ and $f$) such that:
    
    $$\mathbb{E}_n[f(X_{n+1})] = g(X_n)$$
    
    we say that $X_0, X_1, \ldots, X_N$ is a Markov process.

- **Example: Markov Stock Price**

    Consider the stock price evolution:
    $$S_{n+1}(\omega_1 \ldots \omega_{n+1}) = \begin{cases} u \cdot S_n(\omega_1 \ldots \omega_n) & \text{if } \omega_{n+1} = H \\ d \cdot S_n(\omega_1 \ldots \omega_n) & \text{if } \omega_{n+1} = T \end{cases}$$

    So,
    $$\mathbb{E}_n[f(S_{n+1})](\omega_1 \ldots \omega_n) = p \cdot f(u S_n(\omega_1 \ldots \omega_n)) + q \cdot f(d S_n(\omega_1 \ldots \omega_n))$$

    Rewriting, we get:
    $$\mathbb{E}_n[f(S_{n+1})] = g(S_n)$$

    Where:
    $$ g(x) = p \cdot f(ux) + q \cdot f(dx)$$

    Thus, the stock price is Markov.
